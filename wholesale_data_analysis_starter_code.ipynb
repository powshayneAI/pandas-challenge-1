```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']
shrunk_df = df[['unit_price', 'qty', 'subtotal']]
shrunk_df.head(3)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['total_weight'] = df['unit_weight'] * df['qty']

df['shipping_price'] = np.where(df['total_weight'] > 50, df['total_weight'] * 7, df['total_weight'] * 10)

weight_price_df = df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']]
weight_price_df.head(3)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df.rename(columns = {"subtotal":"line_subtotal"}, inplace = True)
sales_tax = ((df['line_subtotal'] + df['shipping_price']) * .0925).round(2)
df['total_price'] = df['line_subtotal'] + df['shipping_price'] + sales_tax
total_price_df = df[['line_subtotal', 'shipping_price', 'total_price']]
total_price_df.head(3)
```

```python
df.rename(columns = { "total_price" : "line_price"}, inplace = True)
df
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
df['line_profit'] = df['line_price'] - df['line_cost']
df.head()
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

summary_data = []
for client_id in c_5_list:
    clients_5_df = df[df['client_id'] == client_id]
    units_purchased_5 = clients_5_df['qty'].sum()
    shipping_price_5 = clients_5_df['shipping_price'].sum()

    
    # revenue_5 = clients_5_df['line_price'].sum() * units_purchased_5
    # cost_5 = (clients_5_df['unit_cost'].sum() * units_purchased_5) + shipping_price_5
    # profit_5 = revenue_5 - cost_5
    revenue_5 = clients_5_df['line_price'].sum()
    cost_5 = clients_5_df['line_cost'].sum()
    profit_5 = clients_5_df['line_profit'].sum()

    summary_data.append({'client_id': client_id,
                                        'total_units' : units_purchased_5,
                                        'total_shipping_price' : shipping_price_5,
                                        'total_revenue' : revenue_5,
                                        'total_cost' : cost_5,
                                        'total_profit' : profit_5
                                       })
summary_df_clients5 = pd.DataFrame(summary_data)
summary_df_clients5
```

```python
# Format the data and rename the columns to names suitable for presentation.
summary_df_clients5.rename(columns={
    'client_id' : 'Client ID',
    'total_units' : 'Units',
    'total_shipping_price' : 'Shipping (millions)',
    'total_revenue' : 'Total Revenue (millions)',
    'total_cost' : 'Total Cost (millions)',
    'total_profit' : 'Total Profit (millions)'
}, inplace = True)
summary_df_clients5
# Define the money columns. 
money_columns = ['Shipping (millions)', 'Total Revenue (millions)', 'Total Cost (millions)', 'Total Profit (millions)']
# Define a function that converts a dollar amount to millions.
def currency_format_millions(amount):
    
    return amount / 1000000

# Apply the currency_format_millions function to only the money columns. 
summary_df_clients5[money_columns] = summary_df_clients5[money_columns].apply(currency_format_millions)
# Rename the columns to reflect the change in the money format. 
summary_df_clients5

```
